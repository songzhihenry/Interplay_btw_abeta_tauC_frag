```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
ctmap = np.loadtxt('./ctmap-0-1.txt')[:,374:]
print(ctmap.shape)
```

```python
figsize = (14,8.46)
fig = plt.figure(figsize=figsize)
grid = plt.GridSpec(12,10)
map = fig.add_subplot(grid[2:, :-2])
bar_top = fig.add_subplot(grid[:2, :-2],sharex=map)
bar_left = fig.add_subplot(grid[2:, -2:],sharey=map)
cmap = cm.Greys
htm = map.imshow(ctmap,cmap=cmap,origin='lower')
bar_top.bar(np.arange(len(ctmap[0])),ctmap.sum(axis=0))
bar_left.barh(np.arange(42),ctmap.sum(axis=1))
bar_top.xaxis.tick_top()
bar_left.xaxis.tick_top()
bar_left.yaxis.tick_right()
cb = plt.colorbar(htm,ax=bar_left,pad=0.2)
map.set_xticks(np.arange(7)*10)
map.set_xticklabels(np.arange(7)*10+375)
map.set_yticks(np.arange(9)*5)
map.set_yticklabels(np.arange(9)*5+1)
```
